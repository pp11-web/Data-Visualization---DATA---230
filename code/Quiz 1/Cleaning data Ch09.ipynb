```python
#Ckecking for the null values in the born_in places
df = df[df.born_in.isnull()]
df.count()
```

```python
#fixing duplicates
dupes_by_names = df[df.duplicated('name')]
dupes_by_names.count()
```

```python
all_dupes = df[df.duplicated('name') | df.duplicated('name', keep='last')]
all_dupes.count()
```

```python
#Filtering the rows thats counts more than 1:
for name, rows in df.groupby('name'):
    if(len(rows)>1):
        print('name: %s, number of rows: %d'%(name, len(rows)))
```
